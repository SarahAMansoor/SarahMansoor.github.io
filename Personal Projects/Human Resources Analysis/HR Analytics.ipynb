### Human Resources Analytics

#### Data Loading and Cleaning

```python
import pandas as pd

# Specify the file path
file_path = 'HR Data.xlsx'

# Read the Excel file into a DataFrame
data = pd.read_excel(file_path)

# Display the first few rows of the DataFrame
data.head()
```

```python
data.info()
```

```python
# Dictionary to map old column names to new column names
column_mapping = {
    'Business Travel': 'Business_Travel',
    'CF_age band': 'CF_Age_Band', 
    'CF_attrition label': 'CF_Attrition_Label', 
    'Education Field': 'Education_Field',
    'emp no': 'Staff_Number', 
    'Employee Number': 'Employee_Number',
    'Job Role': 'Job_Role',
    'Marital Status': 'Marital_Status', 
    'Over Time': 'Over_Time', 
    'Over18': 'Over_18',
    'Training Times Last Year': 'Training_Times_Last_Year', 
    'CF_current Employee': 'CF_Current_Employee',
    'Daily Rate': 'Daily_Rate',
    'Distance From Home': 'Distance_From_Home', 
    'Employee Count': 'Employee_Count', 
    'Environment Satisfaction': 'Environment_Satisfaction',
    'Hourly Rate': 'Hourly_Rate', 
    'Job Involvement': 'Job_Involvement',
    'Job Level': 'Job_Level',
    'Job Satisfaction': 'Job_Satisfaction', 
    'Monthly Income': 'Monthly_Income', 
    'Monthly Rate': 'Monthly_Rate',
    'Num Companies Worked': 'Num_Companies_Worked', 
    'Percent Salary Hike': 'Percent_Salary_Hike',
    'Performance Rating': 'Performance_Rating', 
    'Relationship Satisfaction': 'Relationship_Satisfaction',
    'Standard Hours': 'Standard_Hours',
    'Stock Option Level': 'Stock_Option_Level', 
    'Total Working Years': 'Total_Working_Years', 
    'Work Life Balance': 'Work_Life_Balance',
    'Years At Company': 'Years_At_Company', 
    'Years In Current Role': 'Years_In_Current_Role',
    'Years Since Last Promotion': 'Years_Since_Last_Promotion',
    'Years With Curr Manager': 'Years_With_Curr_Manager'
}

# Rename columns
data = data.rename(columns=column_mapping)

# Display the DataFrame with new column names
data.head()

data.info()
```

```python
# Delete column 'Staff_Number' using del
del data['Staff_Number']
```

```python
# Convert columns to categorical dtype (factor)
data['Attrition'] = data['Attrition'].astype('category')

data['Business_Travel'] = data['Business_Travel'].astype('category')

data['CF_Age_Band'] = data['CF_Age_Band'].astype('category')

data['CF_Attrition_Label'] = data['CF_Attrition_Label'].astype('category')

data['Department'] = data['Department'].astype('category')

data['Education_Field'] = data['Education_Field'].astype('category')

data['Gender'] = data['Gender'].astype('category')

data['Job_Role'] = data['Job_Role'].astype('category')

data['Marital_Status'] = data['Marital_Status'].astype('category')

data['Over_Time'] = data['Over_Time'].astype('category')

data['Over_18'] = data['Over_18'].astype('category')

data['Education'] = data['Education'].astype('category')


# Display the DataFrame with data types
print(data.dtypes)
```

```python
data.info()
```

```python
# List out the categories of each categorical column
for col in data.select_dtypes(include=['category']).columns:
    categories = data[col].cat.categories
    categories_list = list(categories)
    print(f"Categories for {col}: {categories_list}")
```

```python
# Delete column 'Over_18' using del
del data['Over_18']
```

#### Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame(data)
```

```python
# 1. Attrition Rate Visualization
plt.figure(figsize=(8, 6))
sns.countplot(x='Attrition', data=df)
plt.title('Attrition Count')
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.show()
```

This visualization presents the count of employees with and without attrition. A high number of employees leaving the organization can indicate underlying issues such as dissatisfaction with work conditions, lack of career growth opportunities, or poor management practices. Understanding attrition rates helps HR departments identify areas for improvement in employee retention strategies, such as enhancing employee engagement, providing career development programs, or addressing workplace concerns promptly.

With an attrition count of around 200 ("Yes") out of 1200 ("No"), it's evident that the attrition rate is relatively low compared to the total workforce size. However, even though the absolute number of employees leaving the organization may seem relatively small, it's essential to investigate the underlying reasons behind attrition to prevent potential negative impacts on employee morale, productivity, and organizational performance.

In interpreting these attrition counts, HR should delve deeper into the reasons why employees are leaving, particularly those in the "Yes" category. This could involve conducting exit interviews, surveys, or qualitative assessments to understand the factors contributing to attrition, such as job dissatisfaction, lack of career advancement opportunities, work-life balance issues, or cultural mismatches. By identifying these factors, HR can implement targeted strategies to address root causes, improve employee retention, and foster a more positive work environment.

Furthermore, while a low attrition rate may indicate overall satisfaction and stability within the workforce, HR should remain vigilant and proactive in monitoring trends and addressing potential areas of concern. By continuously evaluating and refining retention strategies, providing opportunities for professional growth and development, and promoting a supportive and inclusive workplace culture, organizations can mitigate attrition risks and maintain a motivated and engaged workforce poised for long-term success.

```python
# 2. Distribution of Numerical Features
numerical_features = ['Age', 'Monthly_Income']  # Add more features as needed

plt.figure(figsize=(12, 8))
for feature in numerical_features:
    plt.subplot(1, len(numerical_features), numerical_features.index(feature)+1)
    sns.histplot(df[feature], kde=True)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

Analyzing the distribution of numerical features like age and monthly income can provide insights into the demographic and economic characteristics of the workforce. HR can utilize these insights to tailor recruitment, compensation, and retention strategies to better meet the diverse needs of employees across different age groups and income brackets.

A normal distribution curve for age with a mean around the mid-30s indicates that the majority of employees cluster around this central value, suggesting a balanced distribution of ages within the organization.

From an HR perspective, understanding the age distribution of the workforce is crucial for various strategic initiatives. It allows HR professionals to assess workforce demographics, plan for succession, and tailor HR policies and programs to meet the needs of employees across different age groups. For instance, HR might develop targeted training and development programs to support career growth for younger employees or implement flexible work arrangements to accommodate the needs of older workers nearing retirement. Additionally, recognizing the diverse perspectives and experiences of employees at different life stages contributes to fostering an inclusive and supportive workplace culture.

A right-skewed distribution with a highest bar at $2500, representing over 300 employees, suggests that a significant portion of the workforce earns lower incomes. This could indicate the presence of entry-level or lower-paying positions within the organization, or a concentration of employees in roles with limited earning potential.

The decrease in the frequency of employees earning higher incomes, with only a small number taking home $5000 or more each month, indicates a disparity in income distribution within the workforce. This could reflect differences in job roles, levels of experience, or skill sets, with higher-income earners likely occupying positions with greater responsibility, expertise, or seniority.

From an HR perspective, understanding the distribution of monthly income is critical for assessing compensation structures, addressing pay equity issues, and promoting employee financial well-being. HR professionals may use this information to evaluate the competitiveness of salary packages, conduct salary benchmarking studies, and advocate for fair and equitable compensation practices. Additionally, HR can implement initiatives such as financial literacy programs, employee assistance programs, or performance-based incentives to support employees in achieving their financial goals and improving overall job satisfaction and retention.

```python
# 3. Correlation Analysis
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

Let's delve into the potential implications of these correlations in the context of HR:

Age and Stock Option Level: A positive correlation between age and stock option level suggests that older employees may be more likely to have higher stock option levels. This correlation could be attributed to factors such as longer tenure within the organization, career progression, or participation in stock-based compensation programs. From an HR perspective, this correlation underscores the importance of considering employees' age and tenure when designing stock option plans and rewards systems. It also highlights the role of stock-based incentives in retaining and motivating employees, particularly those with longer-term commitments to the organization.

Total Working Years and Job Level: The positive correlation between total working years and job level indicates that employees with more years of experience tend to occupy higher-level positions within the organization. This correlation is intuitive, as employees typically progress in their careers and gain promotions as they accumulate experience and expertise. HR can leverage this insight to develop career development pathways, succession plans, and talent management strategies that recognize and reward employees' long-term contributions and achievements. Additionally, it underscores the importance of providing opportunities for skill development, mentorship, and advancement to support employees' career growth and retention.

Total Working Years and Monthly Income: The positive correlation between total working years and monthly income suggests that employees with more years of experience tend to earn higher incomes. This correlation reflects common practices in many organizations, where salaries are often tied to seniority and experience levels. From an HR perspective, it highlights the importance of implementing fair and transparent compensation structures that reward employees' tenure, skills, and contributions. It also underscores the need for regular salary reviews, market benchmarking, and performance-based compensation strategies to ensure that employees are fairly compensated and motivated to continue contributing to the organization's success.

In summary, these high correlations provide valuable insights into the relationships between various factors such as age, tenure, job level, and compensation within the workforce. By understanding and leveraging these correlations, HR can make informed decisions and implement targeted strategies to optimize employee engagement, retention, and organizational performance.

```python
# 4. Attrition by Categorical Features
categorical_features = ['Business_Travel', 'Department', 'Education_Field']  # Add more features as needed

plt.figure(figsize=(12, 8))
for feature in categorical_features:
    plt.subplot(1, len(categorical_features), categorical_features.index(feature)+1)
    sns.countplot(x=feature, hue='Attrition', data=df)
    plt.title(f'Attrition by {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Examining attrition rates across categorical features such as business travel frequency or department can reveal patterns and correlations between specific work conditions and employee turnover. Understanding these trends enables HR to address root causes of attrition, foster a supportive work environment, and implement targeted interventions to improve employee retention and job satisfaction.

Business Travel and Attrition: The observation that most employees who travel rarely have a lower attrition rate ("No") suggests that there may be a correlation between travel frequency and job satisfaction or work-life balance. Employees who travel less frequently may perceive their work arrangements as more manageable or conducive to their personal lives, leading to higher job satisfaction and lower turnover. From an HR perspective, understanding the impact of travel on employee well-being and satisfaction can inform policies and practices related to travel requirements, remote work options, and work-life balance initiatives.

Department and Attrition: The finding that most employees in the R&D department have a lower attrition rate ("No") compared to the Sales department aligns with common trends in turnover rates across different job functions or departments. R&D roles may offer opportunities for innovation, intellectual challenge, and career growth, leading to higher job satisfaction and retention among employees in this department. Conversely, the Sales department may involve higher pressure, performance expectations, and turnover due to factors such as sales targets, client relationships, and market competition. HR can use this information to tailor recruitment, retention, and talent development strategies specific to each department's needs and challenges.

Education Field and Attrition: The observation that employees with backgrounds in human resources, medical, and marketing have a lower attrition rate ("No") compared to those with technical degrees highlights the influence of educational background on job satisfaction and retention. Employees with degrees in fields such as human resources, medical, or marketing may have specialized skills or interests that align closely with their job roles, leading to higher job satisfaction and retention. Conversely, employees with technical degrees may experience higher attrition rates due to factors such as career opportunities, industry demand, or alignment between their skills and job roles. HR can leverage this insight to design recruitment strategies, employee development programs, and career pathways that align with employees' educational backgrounds and career aspirations.

```python
# 6. Tenure Analysis

# Histogram showing the distribution of Total_Working_Years
plt.figure(figsize=(8, 6))
sns.histplot(df['Total_Working_Years'], bins=20, kde=False)
plt.title('Distribution of Total Working Years')
plt.xlabel('Total Working Years')
plt.ylabel('Count')
plt.show()
```

```python
# 7. Promotion and Career Growth

# Bar chart showing the distribution of Years_Since_Last_Promotion and Years_In_Current_Role
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Years_Since_Last_Promotion', y='Years_In_Current_Role', hue='Attrition', bins=20, kde=False)
plt.title('Distribution of Years Since Last Promotion vs Years In Current Role')
plt.xlabel('Years Since Last Promotion')
plt.ylabel('Years In Current Role')
plt.legend(title='Attrition')
plt.show()
```

```python
# 8. Comparison of attrition rates for employees who have been promoted recently versus those who haven't
promotion_attrition_counts = df.groupby(['Years_Since_Last_Promotion', 'Attrition']).size().unstack()
promotion_attrition_counts.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Attrition by Years Since Last Promotion')
plt.xlabel('Years Since Last Promotion')
plt.ylabel('Count')
plt.show()
```

```python
# 9. Work-Life Balance

# Distribution of Work_Life_Balance ratings and their relationship with attrition
plt.figure(figsize=(8, 6))
sns.countplot(x='Work_Life_Balance', hue='Attrition', data=df)
plt.title('Attrition by Work Life Balance')
plt.xlabel('Work Life Balance')
plt.ylabel('Count')
plt.show()
```

The distribution of work-life balance ratings and their relationship with attrition provides valuable insights into employees' perceptions of their ability to manage work-related demands and personal responsibilities. HR can use these findings to implement policies and initiatives aimed at promoting work-life balance, such as flexible work arrangements, wellness programs, or stress management resources, thereby enhancing employee well-being and reducing turnover.

```python
# 10. Gender and Diversity Analysis

# Stacked bar chart showing attrition rates by gender
gender_attrition_counts = df.groupby(['Gender', 'Attrition']).size().unstack()
gender_attrition_counts.plot(kind='bar', stacked=True, figsize=(8, 6))
plt.title('Attrition by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

Analyzing attrition rates by gender and diversity metrics offers insights into potential disparities in retention and career progression opportunities among different demographic groups. A stacked bar chart or pie chart showing attrition rates by gender can highlight any gender-based inequities in the workplace, such as gender bias in promotions, pay gaps, or unequal access to leadership roles. Similarly, visualizing diversity indices across departments or job roles can identify areas where diversity and inclusion efforts may be lacking, leading to higher turnover rates among underrepresented groups. HR can leverage these insights to foster a more inclusive and equitable work environment, implement diversity training programs, and establish diversity metrics to track progress over time.
